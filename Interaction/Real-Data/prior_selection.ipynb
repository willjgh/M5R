# Ensembl API

```python
import pandas as pd
import ensembl_rest as ens
import time
```

```python
data_csv = pd.read_csv("./Datasets/GSE151334_FIB_counts_cleaned.csv", index_col=0)
```

```python
data_csv.head()
```

```python
gene_symbols = list(data_csv.index)

biotypes = []
failures = 0
for i, symbol in enumerate(gene_symbols):
    if i % 1000 == 0:
        print(i)

    if i % 14 == 0:
        time.sleep(2)
    try:
        info = ens.symbol_lookup(
            species='homo sapiens',
            symbol=symbol
        )
        biotypes.append(info['biotype'])
    except ens.HTTPError:
        failures += 1
        biotypes.append("Failure")
print(f"{failures} failed lookups")
```

```python
biotypes_dict = {i: btype for i, btype in enumerate(biotypes)}
```

```python
import json
json.dump(biotypes_dict, open("./Datasets/GSE151335_RNA_types_cleaned.json", 'w'))
```

```python
import numpy as np
import matplotlib.pyplot as plt
labels = np.unique(biotypes, return_counts=True)
plt.pie(labels[1], labels=labels[0], labeldistance=1);
```
