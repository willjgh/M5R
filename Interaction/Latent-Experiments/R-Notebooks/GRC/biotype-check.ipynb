## Old version

Ensembl changed from using "HIST" for histone genes to "H" around 2020. R notebooks use "HIST" and were uploaded in 2020 so must use biotypes from before this date and change.

R notebooks also use "lincRNA" type which was retired around 2019 and combined into general "lncRNA". So must use biotypes from before this as well.

Fetch data from latest release before 2019, `may2015` (july 2019 archive unavailable)

```python
o_df = pd.read_csv("GRCh38.id.name.biotype.old.tsv", sep="\t")
o_df = o_df[['external_gene_name', 'gene_biotype']].drop_duplicates().dropna()
o_df = o_df.rename(columns={'external_gene_name': 'name', 'gene_biotype': 'type'})
o_df = o_df.set_index('name')
o_df
```

```python
o_df['type'].unique()
```

```python
[x for x in o_df.index if "HIST" in x] 
```
