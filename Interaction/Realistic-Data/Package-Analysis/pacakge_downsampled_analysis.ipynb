## Correlation change: original vs downsampled

```python
def correlation_change(original_results, downsampled_results, size):

    # correlations before and after downsampling
    corr_original = []
    corr_downsampled = []

    # original correlations         
    for result in original_results:
                    
        # loop over samples
        for i in range(size):
            
            corr_original.append(result.result_dict[i]['statistic'])

    # downsampled correlations
    for result in downsampled_results:    
            
        # loop over samples
        for i in range(size): 

            corr_downsampled.append(result.result_dict[i]['statistic'])

    # scatter plot
    fig, axs = plt.subplots(figsize=(6, 6))
    axs.scatter(corr_original, corr_downsampled)
    axs.set_xlabel("Original correlations")
    axs.set_ylabel("Downsampled correlations")
    plt.show()
```

```python
def correlation_change_colour(original_results, downsampled_results, size):

    from matplotlib.lines import Line2D

    # correlations before and after downsampling
    corr_original = []
    corr_downsampled = []
    cols = []

    # original correlations         
    for result in original_results:
                    
        # loop over samples
        for i in range(size):

            if result.dataset.param_dataset.iloc[i]['k_reg'] == 0.0:
                cols.append("orange")
            else:
                cols.append("blue")
            
            corr_original.append(result.result_dict[i]['statistic'])

    # downsampled correlations
    for result in downsampled_results:    
            
        # loop over samples
        for i in range(size): 

            corr_downsampled.append(result.result_dict[i]['statistic'])

    # scatter plot
    fig, axs = plt.subplots(figsize=(6, 6))
    axs.scatter(corr_original, corr_downsampled, color=cols)
    axs.set_xlabel("Original correlations")
    axs.set_ylabel("Downsampled correlations")
    # legend
    handles = [
        Line2D([0], [0], c="orange", marker='o', linestyle=''),
        Line2D([0], [0], c="blue", marker='o', linestyle='')
    ]
    labels = [
        "Independent sample",
        "Interaction sample"]
    axs.legend(handles, labels)
    plt.show()
```

```python
correlation_change([pearson_one_easy, pearson_one_hard], [pearson_one_easy_downsampled, pearson_one_hard_downsampled], 200)
```

```python
correlation_change_colour([pearson_one_easy, pearson_one_hard], [pearson_one_easy_downsampled, pearson_one_hard_downsampled], 200)
```

## Correlation distribution

```python
def correlation_dist_split(*results, bins_sig=10, bins_not=10):

    # statistics of significant / not results
    stat_sig = []
    stat_not = []

    # loop over all results
    for result in results:

        # get correlation statistics
        for key, val in result.result_dict.items():
            if val['pvalue'] < 0.05:
                stat_sig.append(val['statistic'])
            else:
                stat_not.append(val['statistic'])

    # plot distribution
    fig, axs = plt.subplots(figsize=(6, 6))
    axs.set_title("Correlation test statistics")
    axs.hist(stat_sig, color="green", alpha=0.75, label="Significant", bins=bins_sig)
    axs.hist(stat_not, color="red", alpha=0.75, label="Not sig.", bins=bins_not)
    axs.set_xlabel("Correlation")
    axs.legend()
    plt.show()
```

```python
def correlation_dist(*results, bins=10):

    # statistics
    stat = []

    # loop over all results
    for result in results:

        # get correlation statistics
        for key, val in result.result_dict.items():
            stat.append(val['statistic'])

    # plot distribution
    fig, axs = plt.subplots(figsize=(6, 6))
    axs.set_title("Correlation test statistics")
    axs.hist(stat, bins=bins)
    axs.set_xlabel("Correlation")
    plt.show()
```

```python
def correlation_dist_indep(*results, bins_sig=10, bins_not=10):

    # statistics of significant / not results
    stat_sig = []
    stat_not = []

    # loop over all results
    for result in results:

        # get correlation statistics
        for key, val in result.result_dict.items():

            # check if independent sample
            if result.dataset.param_dataset.iloc[key]['k_reg'] == 0.0:

                if val['pvalue'] < 0.05:
                    stat_sig.append(val['statistic'])
                else:
                    stat_not.append(val['statistic'])

    # plot distribution
    fig, axs = plt.subplots(figsize=(6, 6))
    axs.set_title("Correlation test statistics")
    axs.hist(stat_sig, color="green", alpha=0.75, label="Significant", bins=bins_sig)
    axs.hist(stat_not, color="red", alpha=0.75, label="Not sig.", bins=bins_not)
    axs.set_xlabel("Correlation")
    axs.legend()
    plt.show()
```
