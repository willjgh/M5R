## Results

### 10%

```python
ind_MF_capture_10_5_df_NB = pd.read_csv("Results-capture/ind_MF_capture_10_5.csv", index_col=0)
int_MF_capture_10_5_df_NB = pd.read_csv("Results-capture/int_MF_capture_10_5.csv", index_col=0)
corr_capture_10_5_df_NB = pd.read_csv("Results-capture/corr_capture_10_5.csv", index_col=0)
```

```python
ind_MF_capture_10_5_df_NB.value_counts()
```

```python
mask = (ind_MF_capture_10_5_df_NB == "INFEASIBLE")
plt.scatter(
    corr_capture_10_5_df_NB['MIR4426_c95_OB'][~mask.values.squeeze()],
    corr_capture_10_5_df_NB['MIR4426_c95_AL'][~mask.values.squeeze()],
    color="green",
    s=5,
    alpha=0.25
)
plt.scatter(
    corr_capture_10_5_df_NB['MIR4426_c95_OB'][mask.values.squeeze()],
    corr_capture_10_5_df_NB['MIR4426_c95_AL'][mask.values.squeeze()],
    color="red",
    s=5
)
plt.axline([0, 0], [0.1, 0.1], color="grey")
```

```python
mask = (ind_MF_capture_10_5_df_NB == "INFEASIBLE")
corr = corr_capture_10_5_df_NB['MIR4426_c95_OB'][mask.values.squeeze()]
lb = int_MF_capture_10_5_df_NB['MIR4426_d3_N1000_c95_mc95_HAR_corr_min']
ub = int_MF_capture_10_5_df_NB['MIR4426_d3_N1000_c95_mc95_HAR_corr_max']
mid = (lb + ub) / 2
plt.scatter(
    corr,
    mid,
    color="blue",
    s=5
)
plt.axline([0, 0], [0.1, 0.1], color="red")
```
